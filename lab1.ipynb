```python
#ex1

import numpy as np
import matplotlib.pyplot as plt

# (a) Generarea axei de timp
t_start = 0
t_stop = 0.03
t_step = 0.0005
t = np.arange(t_start, t_stop, t_step)

# (b) Construirea i afiarea semnalelor x(t), y(t), z(t)
x_t = np.cos(520 * np.pi * t + np.pi / 3)
y_t = np.cos(280 * np.pi * t - np.pi / 3)
z_t = np.cos(120 * np.pi * t + np.pi / 3)

# Crearea figurii pentru semnalele continue
plt.figure(figsize=(10, 8))

# x(t)
plt.subplot(3, 1, 1)
plt.plot(t, x_t, label='x(t)')
plt.title('x(t) = cos(520t + /3)')
plt.xlabel('Timp (s)')
plt.ylabel('Amplitudine')
plt.legend()

# y(t)
plt.subplot(3, 1, 2)
plt.plot(t, y_t, label='y(t)')
plt.title('y(t) = cos(280t - /3)')
plt.xlabel('Timp (s)')
plt.ylabel('Amplitudine')
plt.legend()

# z(t)
plt.subplot(3, 1, 3)
plt.plot(t, z_t, label='z(t)')
plt.title('z(t) = cos(120t + /3)')
plt.xlabel('Timp (s)')
plt.ylabel('Amplitudine')
plt.legend()

plt.tight_layout()
plt.show()

# (c) Eantionarea semnalelor la 200 Hz
f_esantionare = 200  # Frecvena de eantionare n Hz
t_e = np.arange(t_start, 1, 1/6)  # Generarea noii axe de timp pentru eantionare

# Calculul semnalelor eantionate
x_n = np.cos(520 * np.pi * t_e + np.pi / 3)
y_n = np.cos(280 * np.pi * t_e - np.pi / 3)
z_n = np.cos(120 * np.pi * t_e + np.pi / 3)

# Crearea figurii pentru semnalele eantionate
fig, axs = plt.subplots(4, 1, figsize=(10, 6))

# x[n]
plt.subplot(311)
plt.plot(t_e, x_n)
plt.stem(t_e, x_n, linefmt='r-', markerfmt='ro', basefmt=' ')
plt.title('x[n] eantionat la 200 Hz')
plt.xlabel('Timp (s)')
plt.ylabel('Amplitudine')

# y[n]
plt.subplot(312)
plt.plot(t_e, y_n)
plt.stem(t_e, y_n, basefmt=' ', linefmt='r-', markerfmt='ro')
plt.title('y[n] eantionat la 200 Hz')
plt.xlabel('Timp (s)')
plt.ylabel('Amplitudine')

# z[n]
plt.subplot(313)
plt.plot(t_e, z_n)
plt.stem(t_e, z_n, linefmt='r-', markerfmt='ro', basefmt=' ')
plt.title('z[n] eantionat la 200 Hz')
plt.xlabel('Timp (s)')
plt.ylabel('Amplitudine')

plt.tight_layout()
plt.show()
```

```python
#ex2

import numpy as np
import matplotlib.pyplot as plt

# (a) Semnal sinusoidal, 400 Hz, 1600 de eantioane
frecventa_a = 400
numar_esantioane_a = 1600
timp_a = np.arange(numar_esantioane_a) / frecventa_a
semnal_a = np.sin(2 * np.pi * frecventa_a * timp_a)

# (b) Semnal sinusoidal, 800 Hz, durata de 3 secunde
frecventa_b = 800
durata_b = 3  # in secunde
timp_b = np.linspace(0, durata_b, int(frecventa_b * durata_b))  # eantionare conform duratei i frecvenei
semnal_b = np.sin(2 * np.pi * frecventa_b * timp_b)

# (c) Semnal de tip "sawtooth", 240 Hz
frecventa_c = 240
durata_c = 3  # arbitrar ales
timp_c = np.linspace(0, durata_c, int(frecventa_c * durata_c))
semnal_c = 2 * (timp_c * frecventa_c - np.floor(0.5 + timp_c * frecventa_c))  # formula pentru semnalul "sawtooth"

# (d) Semnal de tip "square", 300 Hz
frecventa_d = 300
durata_d = 3  # arbitrar ales
timp_d = np.linspace(0, durata_d, int(frecventa_d * durata_d))
semnal_d = np.sign(np.sin(2 * np.pi * frecventa_d * timp_d))  # formula pentru semnalul ptratic

# (e) Semnal 2D aleator
dimensiune_e = (128, 128)
semnal_e = np.random.rand(*dimensiune_e)

# (f) Semnal 2D la alegere
dimensiune_f = (128, 128)
semnal_f = np.zeros(dimensiune_f)
semnal_f[32:-32, 32:-32] = np.ones((64, 64))

# Afiarea graficelor
plt.figure()

# Semnalul (a)
plt.subplot(321)
plt.plot(timp_a, semnal_a)
plt.title('Semnal sinusoidal 400 Hz')

# Semnalul (b)
plt.subplot(322)
plt.plot(timp_b, semnal_b)
plt.title('Semnal sinusoidal 800 Hz')

# Semnalul (c)
plt.subplot(323)
plt.plot(timp_c, semnal_c)
plt.title('Semnal sawtooth 240 Hz')

# Semnalul (d)
plt.subplot(324)
plt.plot(timp_d, semnal_d)
plt.title('Semnal square 300 Hz')

# Semnalul (e)
plt.subplot(325)
plt.imshow(semnal_e, cmap='gray', interpolation='nearest')
plt.title('Semnal 2D aleator')

# Semnalul (f)
plt.subplot(326)
plt.imshow(semnal_f, cmap='gray', interpolation='nearest')
plt.title('Semnal 2D ales')

plt.tight_layout()
plt.show()
```
