```python
#ex1

import numpy as np
import matplotlib.pyplot as plt

N = 8
n = np.arange(N)
k = n.reshape((N, 1))
F_N = np.sqrt(1/N) * np.exp(-2j * np.pi * k * n / N)

product = np.dot(F_N, np.conj(F_N.T))
is_unitary = np.allclose(product, np.eye(N))

print(f"Matricea este unitarÄƒ: {is_unitary}")


for i in range(N):
    plt.subplot(N, 2, 2*i+1)
    plt.plot(n, F_N[i].real)
    plt.title(f"Real - Linia {i+1}")

    plt.subplot(N, 2, 2*i+2)
    plt.plot(n, F_N[i].imag)
    plt.title(f"Imaginar - Linia {i+1}")

    plt.savefig("ex1.pdf")
    plt.savefig("ex1.png")
#plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 8
n = np.arange(N)
k = n.reshape((N, 1))
omega = np.exp(-2j * np.pi / N)

F_NY = np.sqrt(1/N) * omega ** (k * n.T)  # Matricea Fourier

# Verificare dacÄƒ F_NY este unitarÄƒ
product = np.dot(F_NY, np.conj(F_NY.T))
is_unitary = np.allclose(product, np.eye(N))

print(f"Matricea este unitarÄƒ: {is_unitary}")

for i in range(N):
    plt.subplot(N, 2, 2*i+1)
    plt.plot(n, F_NY[i].real)
    plt.title(f"Real - Linia {i+1}")

    plt.subplot(N, 2, 2*i+2)
    plt.plot(n, F_NY[i].imag)
    plt.title(f"Imaginar - Linia {i+1}")

plt.tight_layout()
plt.savefig("ex1.pdf")
plt.savefig("ex1.png")
plt.show()
```
