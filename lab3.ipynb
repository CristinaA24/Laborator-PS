```python
#ex1

import numpy as np
import matplotlib.pyplot as plt

N = 8
n = np.arange(N)
k = n.reshape((N, 1))
F_N = np.sqrt(1/N) * np.exp(-2j * np.pi * k * n / N)

product = np.dot(F_N, np.conj(F_N.T))
is_unitary = np.allclose(product, np.eye(N))

print(f"Matricea este unitarÄƒ: {is_unitary}")


for i in range(N):
    plt.subplot(N, 2, 2*i+1)
    plt.plot(n, F_N[i].real)
    plt.title(f"Real - Linia {i+1}")

    plt.subplot(N, 2, 2*i+2)
    plt.plot(n, F_N[i].imag)
    plt.title(f"Imaginar - Linia {i+1}")

    plt.savefig("ex1.pdf")
    plt.savefig("ex1.png")
#plt.tight_layout()
plt.show()
```

```python
#ex2

import numpy as np
import matplotlib.pyplot as plt
import math


sin_function = lambda x: np.sin(2 * np.pi * 7 * x)

fig, axs = plt.subplots(6, figsize=(5, 10))
t = np.linspace(0, 1, 1000)
y = sin_function(t)

axs[0].plot(t, y)
axs[0].set_title("Figura 1.1")

functie1 = math.e ** (-2 * np.pi * 1j * t * 1)
axs[1].plot(y * functie1.real, y * functie1.imag)
axs[1].set_title("Figurile 1.2 si 2.1")

functie2 = math.e ** (-2 * np.pi * 1j * t * 2)
axs[2].plot(y * functie2.real, y * functie2.imag)
axs[2].set_title("Figura 2.1")

functie3 = math.e ** (-2 * np.pi * 1j * t * 5)
axs[3].plot(y * functie3.real, y * functie3.imag)
axs[3].set_title("Figura 2.2")

functie4 = math.e ** (-2 * np.pi * 1j * t * 7)
axs[4].plot(y * functie4.real, y * functie4.imag)
axs[4].set_title("Figura 2.3")

functie5 = math.e ** (-2 * np.pi * 1j * t * 10)
axs[5].plot(y * functie5.real, y * functie5.imag)
axs[5].set_title("Figura 2.4")

plt.savefig("ex2.pdf")
plt.savefig("ex2.png")

plt.show()
```
