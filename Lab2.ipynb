```python
#ex1

import numpy as np
import matplotlib.pyplot as plt

# Parametrii semnalului
amplitudine = 1.0      # Amplitudinea semnalului
frecventa = 5.0        # Frecvența semnalului în Hz
faza = np.pi / 2       # np.pi/2 va face cele două semnale identice
timp_total = 2.0       # Durata semnalului în secunde

# Generarea semnalului
timp = np.linspace(0, timp_total, int(250 * timp_total))  # 250 de eșantioane pe secundă
semnal_sinus = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza)
semnal_cosinus = amplitudine * np.cos(2 * np.pi * frecventa * timp)

# Crearea subplot-urilor
fig, axs = plt.subplots(2, 1, figsize=(10, 6))

# Primul subplot cu semnalul sinus
axs[0].plot(timp, semnal_sinus, 'b-')
axs[0].set_title('Semnal Sinusoidal')
axs[0].set_xlabel('Timp (s)')
axs[0].set_ylabel('Amplitudine')

# Al doilea subplot cu semnalul cosinus
axs[1].plot(timp, semnal_cosinus, 'r-')
axs[1].set_title('Semnal Cosinusoidal')
axs[1].set_xlabel('Timp (s)')
axs[1].set_ylabel('Amplitudine')

# Afișarea graficelor
plt.tight_layout()
plt.show()
```

```python
#ex2
import numpy as np
import matplotlib.pyplot as plt
# Parametrii semnalului
amplitudine = 1.0      # Amplitudinea semnalului
frecventa = 5.0
timp_total = 1.0  # Frecvența semnalului în Hz
faza1 = np.pi / 2       # np.pi/2 va face cele două semnale identice
faza2 = np.pi / 3
faza3 = np.pi
faza4 = np.pi / 4
valori = 500
timp = np.linspace(0, timp_total, int(500 * timp_total))  # 500 de eșantioane pe secundă

functieZN = np.random.normal(size = (valori, 1))
normaZN = np.linalg.norm(functieZN)
print(normaZN)

semnal_sinus1 = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza1)
semnal_sinus2 = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza2)
semnal_sinus3 = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza3)
semnal_sinus4 = amplitudine * np.sin(2 * np.pi * frecventa * timp + faza4)

normaXN1 = np.linalg.norm(semnal_sinus1)
normaXN2 = np.linalg.norm(semnal_sinus2)
normaXN3 = np.linalg.norm(semnal_sinus3)
normaXN4 = np.linalg.norm(semnal_sinus4)

gama1 = np.sqrt(normaXN1  ** 2 / (0.1 * normaZN **  2))
gama2 = np.sqrt(normaXN2 ** 2/ (1 * normaZN ** 2))
gama3 = np.sqrt(normaXN3 ** 2/ (10 * normaZN ** 2))
gama4 = np.sqrt(normaXN4 ** 2/ (100 * normaZN ** 2))


semnal_sinus_zgomot1 = semnal_sinus1 + gama1 * functieZN
semnal_sinus_zgomot2 = semnal_sinus2 + gama2 * functieZN
semnal_sinus_zgomot3 = semnal_sinus3 + gama3 * functieZN
semnal_sinus_zgomot4 = semnal_sinus4 + gama4 * functieZN


# Crearea subplot-urilor
fig, axs = plt.subplots(4, 1, figsize=(10, 6))

# Primul subplot cu faza1
#axs[0].plot(timp, semnal_sinus_zgomot1, 'b-')
axs[0].plot(timp, semnal_sinus1, 'o-')
axs[0].set_title('Semnal faza1')
axs[0].set_xlabel('Timp (s)')
axs[0].set_ylabel('Amplitudine')

# Al doilea subplot cu faza2
#axs[1].plot(timp, semnal_sinus_zgomot2, 'r-')
axs[1].plot(timp, semnal_sinus2, 'o-')
axs[1].set_title('Semnal faza2')
axs[1].set_xlabel('Timp (s)')
axs[1].set_ylabel('Amplitudine')

# Al doilea subplot cu faza3
#axs[2].plot(timp, semnal_sinus_zgomot3, 'y-')
axs[2].plot(timp, semnal_sinus3, 'o-')
axs[2].set_title('Semnal faza3')
axs[2].set_xlabel('Timp (s)')
axs[2].set_ylabel('Amplitudine')

# Al doilea subplot cu faza4
#axs[3].plot(timp, semnal_sinus_zgomot4, 'g-')
axs[3].plot(timp, semnal_sinus4, 'o-')
axs[3].set_title('Semnal faza4')
axs[3].set_xlabel('Timp (s)')
axs[3].set_ylabel('Amplitudine')
```
